[Makie.jl](https://docs.makie.org/stable/)

````julia
using CairoMakie
````

````julia
f = Figure(backgroundcolor = :tomato)
````

````julia
f = Figure(backgroundcolor = :tomato, size = (800, 300))
````

````julia
f = Figure()
ax = Axis(f[1, 1])
f
````

````julia
f = Figure()
ax = Axis(f[1, 1],
    title = "A Makie Axis",
    xlabel = "The x label",
    ylabel = "The y label"
)
f
````

````julia
f = Figure()
ax = Axis(f[1, 1])
x = range(0, 10, length=100)
y = sin.(x)
lines!(ax, x, y)
f
````

````julia
f = Figure()
ax = Axis(f[1, 1])
x = range(0, 10, length=100)
y = sin.(x)
scatter!(ax, x, y)
f
````

````julia
x = LinRange(0, 10, 100)
y = sin.(x)

fig = Figure()
ax1 = Axis(fig[1, 1])
ax2 = Axis(fig[1, 2])
ax3 = Axis(fig[2, 1:2])

fig
````

````julia
lines!(ax1, x, y, color = :red)
lines!(ax2, x, y, color = :blue)
lines!(ax3, x, y, color = :green)

fig
````
